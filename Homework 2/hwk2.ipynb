```python
import pandas as pd
```

```python
name = ["Vivian Baskette","Jamison Marney", "Marie Mulero", "Trudi Kimmell", "Stephanie Lindemann", "Dia Werely", "Abbie Lame", "Misti Luce", "Wilda Byerly"]
age = [1, 25, 27, 30, 32, 35, 40, 41, 1000]
tax = [7, 11, 8, 13, 7, 8, 9, 10, 12]
income= [70, 110, 80, 130, 120, 80, 90, 100, 120]

label = ['t1', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9']

df = pd.DataFrame({"Name": name, "Age": age, "Income": income, "Tax": tax})
df.index = label

print(df)
```

```python
df.hist(column="Age")
df.hist(column="Income")
```

```python
mean = df["Age"].mean()
print(mean)
std = df["Age"].std()
print(std)

z_score = []
for i in df["Age"]:
    x = (i - mean) / std
    z_score.append(x)
print(z_score)
```
